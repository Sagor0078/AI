```python
def Transfrom_data(data):
  #Encoding label into number
  encoding_data = ({'NEUTRAL' : 0,'POSITIVE' : 1,'NEGATIVE' : 2})
  data_encoded = data.replace(encoding_data)
  #getting brain siognal into x variable
  x = data_encoded.drop(["label"],axis = 1)
  #getting label into y variable
  y = data_encoded.loc[:,'label'].values
  scaler = StandardScaler()
  #Scaling Brain Signals
  scaler.fit(x)
  X = scaler.transform(x)
  Y = to_categorical(y)
  return X,Y
```

```python
X,Y = Transfrom_data(data)
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size = 0.2,random_state = 4)
```

```python
#checkig shape of dataset
xtrain.shape[1]
```

```python
def create_model():
  #input layer of model brain signal 
  inputs = tf.keras.Input(shape=(xtrain.shape[1],))
  #Hidden Layer for Brain signal using LSTM(GRU)
  expand_dims = tf.expand_dims(inputs,axis = 2)
  gru = tf.keras.layers.GRU(256,return_sequences=True) (expand_dims)
  #flattern vector layer into vector form (one dimentional array)
  flatten = tf.keras.layers.Flatten()(gru)
  #output layer of model
  outputs = tf.keras.layers.Dense(3,activation = 'softmax') (flatten)

  model = tf.keras.Model(inputs = inputs,outputs = outputs)
  print(model.summary())
  return model
```

```python
#creating model
lstmmodel = create_model()
#Compiling model
lstmmodel.compile(
    optimizer = 'adam',
    loss = 'categorical_crossentropy',
    metrics = ['accuracy']
)
```

```python
history = lstmmodel.fit(xtrain,ytrain,epochs = 10,validation_split = 0.1)
loss , acc = lstmmodel.evaluate(xtest,ytest)
```

```python
  # Loss and Accuracy of model on Testing Dataset
print(f"Loss on Testing : {loss*100}",f"\nAccuracy on Training : {acc*100}")
```

```python
pred = lstmmodel.predict(xtest)
```

```python
# create function for confusion matrix
def plot_confussion_matrix(cm,names,title = 'Confusion matrix',cmap = plt.cm.Blues):
  plt.imshow(cm,interpolation = 'nearest',cmap=cmap)
  plt.title(title)
  plt.colorbar()
  tick_marks = np.arange(len(data.label.unique()))
  plt.xticks(tick_marks,names,rotation = 90)
  plt.yticks(tick_marks,names)
  plt.tight_layout()
  plt.ylabel('True label')
  plt.xlabel('Predicted label')
```

```python
#after getting prediction checking maximium score prediction to claim which emoton this brain signal belongs to
pred1 = np.argmax(pred,axis = 1)
```
